```python
import pandas as pd
import matplotlib.pyplot as plt

# Load data
file_path = r"C:\Users\ASUS\AppData\Roaming\Python\Python313\site-packages\pandas\io\excel\data harian.xlsx"  # Replace with your actual file path
data = pd.read_excel(file_path)
data['tanggal'] = pd.to_datetime(data['tanggal'])

# Plotting
plt.figure(figsize=(12, 6))
plt.plot(data['tanggal'], data['tone'], color='#006994', linewidth=1)

# Buat label
plt.title("Tone Pemberitaan Kontan.co.id Januari 2019 - Juni 2024")
plt.xlabel("Quarterly Time Periods/Periode Waktu Kuartalan")
plt.ylabel("Tone")
plt.yticks(range(-40, 41, 10))  # Set y-axis range from -40 to 40
plt.ylim(-40, 40)

# Buat label quarter pada x-axis
quarter_dates = pd.date_range(start="2019-01-01", end="2024-06-30", freq='Q')
quarter_labels = [f'Q{q} {y}' for y in range(2019, 2025) for q in range(1, 5)][:len(quarter_dates)]
plt.xticks(ticks=quarter_dates, labels=quarter_labels, rotation=45)

# Tambahan note
plt.figtext(0.1, 0.01, "*data diambil berdasarkan kata kunci khusus", ha="left", fontsize=7)

# Save
plt.savefig("tone_pemberitaan_kontan_data harian.png", dpi=300, bbox_inches='tight')  # dpi=300 for HD quality

plt.grid()
plt.tight_layout()
plt.show()
```
